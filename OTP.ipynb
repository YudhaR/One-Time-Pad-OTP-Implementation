```python
import string
import math
import random

def vigenere_encrypt():
    plaintext = input("Masukkan plain text: ").upper()

    # Membuat kunci acak dengan panjang yang sama seperti plaintext
    key = ''.join(random.choice(string.ascii_uppercase) for _ in range(len(plaintext)))

    # list yang berisikan semua 26 karakter
    all_letters= string.ascii_uppercase #26

    # variable hasil cipher text
    cipher_txt=[]

    # perhitungan karakter chiper dalam loop for
    for i in range(len(plaintext)):
        cipher_txt.append(all_letters[(ord(plaintext[i]) + ord(key[i])) % len(all_letters)])

    cipher_txt= "".join(cipher_txt)

    print("Kode OTP:", key)

    # pengembalian hasil cipher text
    return cipher_txt

def vigenere_decript():
    msg = input("Masukkan Cipher text : ").upper()
    key = input("masukkan kunci : ")

    # list yang berisikan semua 26 karakter
    all_letters= string.ascii_uppercase #26

    # variable hasil plain text
    decrypt_txt = []

    # pengurangan ascii karakter dengan kunci
    for i in range(len(msg)):
        decrypt_txt.append(all_letters[(ord(msg[i]) - ord(key[i])) % len(all_letters)])

    decrypt_txt= "".join(decrypt_txt)

    # mengembalikan Hasil decrypt
    return decrypt_txt


def main():
    while (1):
        choice = int(
            input("\n----------------\nImplementasi OTP pada Algoritma Vigenere\n----------------\n1. Enkripsi \n2. Dekripsi \n3. Keluar \nPilihan: "))
        if choice == 1:
            print("----------------")
            # memanggil function transposisi dengan parameter substitusi
            # function substitusi akan dieksekusi terlebih dahulu, untuk mendapatkan nilai yang akan digunakan sebagai parameter transposisi
            print("Ciphertext : ",vigenere_encrypt())
        elif choice == 2:
            print("----------------")
            # memanggil function substitusi dengan parameter transposisi
            # function transposisi akan dieksekusi terlebih dahulu, untuk mendapatkan nilai yang akan digunakan sebagai parameter substitusi
            print("Plain text : ",vigenere_decript())

        elif choice == 3:
            exit()
        else:
            print("\nChoose correct choice: ")


if __name__ == "__main__":
    main()
```
